```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
#gerando uma massa de dados
x,y = make_regression(n_samples=200,n_features=1,noise=20)
#gerando grafico
plt.scatter(x,y)
plt.show()
```

```python
modelo.fit(x,y)
```
